```python
import numpy as np
from zombies import d_zombie, plot_it, zombie_human_behavior
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
y_0 = np.array([100,50,0,0])
solution = solve_ivp(
    d_zombie,
    (0, 10),
    y_0,
    first_step=0.05,
    max_step=0.05,
    args=(0.25,0.1,0.5, 1, 0.5)
)
plot_it(solution)
```

```python
y_0 = [99, 10, 0, 0, 0, 0.5, 0.5, 0]
params = np.random.uniform(size=(14,))
params = [0.65, 0.0185, 0.0078, 0.9, 0.01, 0.45] + list(np.random.uniform(size=(8,)))
solution = solve_ivp(
    zombie_human_behavior,
    (0, 10),
    y_0,
    first_step=0.05,
    max_step=0.05,
    args=params
)
plt.plot(solution.t, solution.y.T[:,[0,1,2]])
plt.xlabel('t')
plt.legend(['Susceptibles', 'Zombies', 'Exposed'], shadow=True, loc='upper right')
plt.title('Population Changes Over Time')
plt.show()
```
